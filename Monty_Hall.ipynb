```
import seaborn as sns
```

```
import pandas as pd
df = pd.read_csv("http://rulelaw.net/downloads/rol-scores.csv")
```

```
# %matplotlib inline
sns.pairplot(df[["RoLScore", "pol_plur", "free_expr", "assoc_org", "per_auto"]])
```

```
import statsmodels.formula.api as smf
regression = smf.ols('RoLScore ~ assoc_org + per_auto + hprop', data=df).fit()
print(regression.summary())
```

```
#strategy: does not switch 
import random
doors = ["door1", "door2", "door3"]
result = 0
i = 0
times = 10000 #this can change to 1000 etc 
while i < times:
  good_door = random.choice(doors)
  first_selection = random.choice(doors)
  other_doors = doors.remove(good_door)
  remaining_doors = doors
  host_selection = random.choice(doors) 
  second_selection = first_selection
  if second_selection == good_door:
    result = result + 1
  i = i + 1  
  doors = ["door1", "door2", "door3"]

print(result)


```
